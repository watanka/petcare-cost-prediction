```python
# 한 마리당 연령대별 추적 데이터
import random
from datetime import datetime, timedelta

def validate_selected_disease(disease_history, disease_name, gender):
    if disease_name in disease_history:
        return False
    if disease_name == '자궁축농증' and gender == '남자':
        return False
    return True

def pick_disease(disease_list, disease_history, gender):
    while True:
        selected_disease = random.choice(disease_list)
        if validate_selected_disease(disease_history, selected_disease, gender):
            break
        
            
    return selected_disease

def random_date(start_date, end_date):
    delta = end_date - start_date
    random_days = random.randint(0, delta.days)
    return start_date + timedelta(days=random_days)

def random_weight(max_weight, age, gender): 
    # 남자면 더 무거움
    # 1살까지 성장. 이후로는 몸무게 유지
    # 8살이후로는 노년기
    percentage = 1.1 if gender == '남자' else 1.0
    if age <= 1:
        weight = max_weight // 2
    elif age <= 8:
        weight = max_weight + random.randint(-max_weight // 3, max_weight // 3)
    else:
        weight /= 0.9
    
    return (weight * percentage * 100) // 100 # 소숫점 두자리까지
        
    

def simulate_disease_history(id: int):
    '''한마리에 대한 질병기록을 생성한다.'''
    disease_history = []
    records = []
    
    breed = random.choice(breeds)
    gender = random.choice(['남자', '여자'])
    age = 1
    max_age = int(breed_info[breed_info['name']==breed]['lifespan'].values[0].split('~')[1])
    max_weight_kg = breed_info[breed_info['name']==breed]['무게'].values[0]
    is_neuter = random.choice(['y', 'n'])
    
    base_record = {
                    'pet_id': id,
                    'breed': breed,
                    'gender': gender,
                    'pet_breed_id': breed_dict[breed],
                    'neuter_yn': is_neuter,
        }
    
    
    
    
    for current_age in range(age, max_age + 1):
        weight_kg = random_weight(max_weight_kg, age, gender)
        for _ in range(random.randint(1, 5)):        
                
            selected_disease = pick_disease(list(disease_dic.keys()), disease_history, gender)
            min_cost, max_cost, prob_trend_by_age, prob = disease_dic[selected_disease]
            prob_by_age = prob_trend_by_age * (current_age/max_age) * prob

            if random.random() <= prob_by_age:
                cost = int(random.uniform(min_cost, max_cost) * prob_trend_by_age // 100 * 100)
                cost += cost * 1.0008 if not is_neuter else 0

                # created_at - birth = age
                created_at = random_date(datetime(2021,1,1,0,0,0), datetime(2024, 5, 27,0,0,0))
                birth = created_at - timedelta(days=(current_age * 12 + random.randint(-7, 12)) * 30)
                
                new_record = base_record | {'disease_name': selected_disease,
                                            'claim_price':  cost, 
                                            'age': current_age,
                                            'created_at': created_at,
                                            'birth': birth.strftime("%Y-%m-%d"),
                                            'accident_content': '',
                                            'weight_kg': weight_kg
                                            }
                records.append(new_record)
                disease_history.append(selected_disease)
                
    return records
        
```

```python
# insurance 기록에서 평균값 내는 코드가 필요함.
class Statistics:
    def __init__(self, df):
        self.df = df.groupby(['pet_breed_id', 'age'])
        self.price_groupby_breed_and_age = self.df['claim_price'].mean()
        self.most_common_diseases_groupby_breed_and_age = self.df['disease_name'].agg(lambda x: x.mode()[0])
        
    
    def to_dict(self, df: pd.DataFrame):
        return df.to_dict()
    
    def average_price_change_by_breed_and_age(self, breed):
        return self.to_dict(self.price_groupby_breed_and_age[breed])
        
    def most_common_disease_by_breed_and_age(self, breed):
        return self.to_dict(self.most_common_diseases_groupby_breed_and_age[breed])
      
    def aggregate_stat(self, breed):
        avg_price = self.average_price_change_by_breed_and_age(breed)
        most_common_disease = self.most_common_disease_by_breed_and_age(breed)
        
        stat = []
        for age, avg_price, disease in zip(avg_price.keys(), avg_price.values(), most_common_disease.values()):
            stat.append({'age': age, 'average_cost' : avg_price, 'disease': disease})
        return stat
        
    
```

```python
from abc import ABC, abstractmethod
from dataclasses import dataclass
from typing import Optional
import smtplib, ssl
from email.mime.text import MIMEText


class Sender(ABC):
    @abstractmethod
    def send(self):
        raise NotImplementedError

@dataclass
class Receiver:
    email: Optional[str] = None
    phone_number: Optional[str] = None
    
    def validate_necessary_info(self, sender: Sender):
        if not getattr(self, sender.type):
            raise ValueError(f"{sender.type} 정보가 없습니다.")
            
            
  
        
        
class EmailSender(Sender):
    type = 'email'
    def __init__(self, smtp_server: str, port: int, email: str, password: str):
        self.smtp_server = smtp_server
        self.port = port
        self.email = email
        self.password = password
        self.context = ssl.create_default_context()

    def send(self, message: str, receiver: Receiver):
        receiver.validate_necessary_info(self)
        
        message = MIMEText(message)
        message['Subject'] = 'sample'
        message['From'] = self.email
        message['To'] = receiver.email
        
        with smtplib.SMTP(self.smtp_server, self.port) as server:
            server.starttls()
            server.login(self.email, self.password)
            server.sendmail(self.email, receiver.email, message.as_string())
    
            
            
            
```
